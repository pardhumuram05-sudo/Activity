```python
# import modules
import nltk
import string
nltk.download('punkt_tab')

def remove_punctuations_func(text):
    # Tokenize the text into individual words
    words = nltk.word_tokenize(text)
    # Remove punctuations from each word
    words_without_punctuations = [word for word in words if word not in string.punctuation]
    # Join the words back into a string
    text_without_punctuations = " ".join(words_without_punctuations)

    return text_without_punctuations

# Test the function
input_text = "Furthermore, I found the support really great. Paid about 180$ for it (5.7inch version, 4.8'' screen)."
result_text = remove_punctuations_func(input_text)
print("OUTPUT REMOVAL OF PUNCTUATIONS IN THE TEXT USING NLTK")
print("Original string:", input_text)
print("Modified string:", result_text)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stop_words(string):
    stop_words = set(stopwords.words('english'))
    words = string.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    new_string = ' '.join(filtered_words)
    return new_string

# Example usage
input_string = "This is an example sentence to remove stop words from."
result_text = remove_stop_words(input_string)
print("Original string:", input_string)
print("Modified string:", result_text)
```

```python
import spacy
def remove_stop_words(string):
    # Load the spaCy English language model
    nlp = spacy.load('en_core_web_sm')

    # Tokenize the string into individual words
    doc = nlp(string)

    # Filter out stop words
    filtered_words = [token.text for token in doc if not token.is_stop]

    # Join the filtered words back into a string
    new_string = ' '.join(filtered_words)

    return new_string

# Example usage
#input_string = "This is an example sentence to remove stop words from."
input_string = "This is a Demo Text for NLP using NLTK. Full form of NLTK is Natural Language Toolkit."
result_text = remove_stop_words(input_string)
print("Original string:", input_string)
print("Modified string:", result_text)
```

```python
# import modules
import unicodedata

def remove_accented_chars_func(text):
    cleaned_accented_chars = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    return cleaned_accented_chars

# Test the function
input_text = "Sómě Áccěntěd těxt and words like résumé, café or exposé."
result_text = remove_accented_chars_func(input_text)
print("OUTPUT REMOVAL OF ACCENTED CHARACTERS IN THE TEXT")
print("Original string:", input_text)
print("Modified string:", result_text)
```

```python
from collections import Counter
def remdup(input):
    input = input.split(" ")
    unqwords = Counter(input)
    s = " ".join(unqwords.keys())
    print("The unique words are:")
    print(list(unqwords.keys()))
    print("Text with no duplicate words:")
    return s

# input text
input_text = "The only thing we have to fear is fear itself"
new_text = remdup(input_text)
print(new_text)
```

```python
import nltk

def convert_lowercase_tokens_func(text):
    # tokenize the text
    tokens = nltk.word_tokenize(text)
    # lowercase the tokens
    lowercased_tokens = [token.lower() for token in tokens]
    return lowercased_tokens

# Test the function
input_text = "I HIGHLY RECOMMEND TO STUDY NATURAL LANGUAGE PROCESSING SUBJECT BEFORE DOING FINAL YEAR PROJECTS !!."
result_text = convert_lowercase_tokens_func(input_text)
print("OUTPUT CONVERSIONS OF LOWER CASE TEXT TOKENS USING NLTK")
print("Original string:", input_text)
print("Modified string:", result_text)
```

```python
import nltk
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

from nltk import word_tokenize, pos_tag, ne_chunk

# Sample text
text = "Barack Obama was the 44th President of the United States and was born in Hawaii."

# Step 1: Tokenize the text
tokens = word_tokenize(text)

# Step 2: Part-of-Speech Tagging
pos_tags = pos_tag(tokens)

# Step 3: Named Entity Recognition
ner_tree = ne_chunk(pos_tags)

# Print the NER tree
print(ner_tree)
```

```python
#Extracting Named Entities from the Tree

def extract_named_entities(tree):
    entities = []
    for subtree in tree:
        if hasattr(subtree, 'label') and subtree.label() in ['PERSON', 'GPE', 'ORGANIZATION']:
            entity = " ".join([leaf[0] for leaf in subtree.leaves()])
            entities.append((entity, subtree.label()))
    return entities

named_entities = extract_named_entities(ner_tree)
print(named_entities)
```
